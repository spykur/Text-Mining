```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
def get_winners():
    Wikipedia = "https://en.wikipedia.org"
    Wikipedia_Nagrody = "https://en.wikipedia.org/wiki/Academy_Award_for_Best_Picture"
    
    response = requests.get(Wikipedia_Nagrody, headers={"User-Agent": "Project"})
    response.raise_for_status()
    soup = BeautifulSoup(response.text, "html.parser")
    
    tables = soup.find_all("table", class_="wikitable")

    winners = []

    for table in tables:
        rows = table.find_all("tr")
        year = None

        for row in rows:
            cells = row.find_all(["th", "td"])

            if len(cells) < 2:
                continue

            if cells[0].name == "th":
                year = cells[0].get_text(strip=True)

            bold = row.find("b")
            if bold:
                link = bold.find("a")
                if link and link.get("href"):
                    title = link.get_text(strip=True)
                    film_url = Wikipedia + link.get("href")

                    winners.append({
                        "title": title,
                        "year": year,
                        "url": film_url
                    })

    return winners
```

Używam BS abu pobrać zawartośc strony na wikipedii. Następnie przeszukuje wszystkie wiersze tabelii w poszukiwaniu nagłówków o danych.
Pomijam wiersze w których jest za mało danych. Zapamiętuje rok aktualnego wierszu, pobieram dane zwycięzcy biorąc pogrubione dane. Zapisuje tytuł rok i url do wcześniej przygotowanej tabeli.

```python
def get_film_details(url):
    
    response = requests.get(url, headers={"User-Agent": "Project"})
    response.raise_for_status()
    soup = BeautifulSoup(response.text, "html.parser")

    rezyser = " "
    infobox = soup.find("table", class_="infobox")
    if infobox:
        rows = infobox.find_all("tr")
        for row in rows:
            header = row.find("th")
            if header and "Directed by" in header.get_text():
                rezyser_dane = row.find("td")
                if rezyser_dane:
                    rezyser = rezyser_dane.get_text(strip=True)
                    break

    fabula = " "
    paragraphs = soup.find_all("p")
    for p in paragraphs:
        text = p.get_text(strip=True)
        if text:
            fabula = text
            break

    return rezyser, fabula
```

Przechodze na stronę wcześniej pobraną dla każdego z filmów. Szukam infobox, następnie w nim danych reżysera i zapisuje je.
Zapisuje pierwszy rozdział jako opis fabuły. 

```python
def save_to_file(winners):
    i=1
    max_i=len(winners)
    with open("wyniki.txt", "w", encoding="utf-8") as f:
        for film in winners:

            rezyser, fabula = get_film_details(film["url"])

            f.write(f"Tytuł: {film['title']}\n")
            f.write(f"Rok: {film['year']}\n")
            f.write(f"Reżyser: {rezyser}\n")
            f.write(f"Opis: {fabula}\n")
            f.write("\n\n")
            
            txt=f"Zapisano dane dla {i}/{max_i} filmu"
            print(txt)
            i=i+1

            time.sleep(3)  
```

Dla każdego filmu który wygrał, przechodzę na jego podstronę i zapisuje reżysera i fabułe. Zapisuje wszystkie informacje po kolei do pliku tekstowego.
Pomiędzy każdym wykoaniem pętli czekam 3 sekundy aby nie mieć problemu z serwerami. Wypisywanie postępu, aby użytkownik mógł ocenić jak długo potrwa program oraz że nadal działa poprawnie.

```python
winners = get_winners()
save_to_file(winners)
```

Funkcja winners pobiera wszystkich zwycięzców oraz zapisuje ich tytuł, rok o raz url do podstrony. 
Funkcja save_to_file uzupełnia dane o reżysera i fabułe pobierając informacje z podstrony oraz zapisuje dane do pliku tekstowego.
